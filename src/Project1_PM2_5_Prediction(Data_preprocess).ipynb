```python
import csv
filename = '/content/drive/Shareddrives/DS/PM2.5_kaggle 2021/Songkhla/Test/Songkhla.txt'
f = open(filename, 'r')
i = 0

with open(filename[:-3]+'csv','w', newline='') as csvfile:
  csv_writer = csv.writer(csvfile, delimiter=',')
  for line in f:
    if i == 9:
      line = line[1:].strip().split(', ')
      csv_writer.writerow(line)
    elif i > 9:
      line = line.strip().split()
      csv_writer.writerow(line)
    i+=1
  csvfile.close()
f.close()
```
